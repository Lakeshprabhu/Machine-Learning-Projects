```python
import pandas as pd

df = pd.read_csv("Melbourne_housing_FULL.csv")

df
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
columd = ['Bedroom2','Bathroom',"Propertycount","Car"]

df[columd] = df[columd].fillna(0)
```

```python
df.isna().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Distance'] = df['Distance'].fillna(0)
```

```python
df['Price'] = df['Price'].fillna(df['Price'].mean())
df['Landsize'] = df['Landsize'].fillna(df['Landsize'].mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].mean())
```

```python
df.isna().sum()
```

```python
df = df.drop(['Lattitude','Longtitude','Date','Postcode','YearBuilt','CouncilArea'],axis='columns')
```

```python
df.isna().sum()
```

```python
df = df.fillna(0)
```

```python
df.isna().sum()
```

```python
df
```

```python
dataset = pd.get_dummies(df,drop_first=True,dtype=int)
```

```python
dataset
```

```python
X = dataset.drop('Price',axis='columns')
```

```python
y = dataset.Price
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2 )
```

```python
from sklearn.linear_model import LinearRegression

mod = LinearRegression()

mod.fit(X_train,y_train)
```
