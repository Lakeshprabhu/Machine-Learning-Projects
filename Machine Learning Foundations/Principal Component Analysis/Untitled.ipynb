```python
df = pd.read_csv("heart.csv")

```

```python
df.describe()
```

```python
df = df[ df['Cholesterol'] <= (df['Cholesterol'].mean() + 3*df['Cholesterol'].std())]

```

```python
df = df[ df['Cholesterol'] >= (df['Cholesterol'].mean() - 3*df['Cholesterol'].std())]
```

```python
df = df[ df['RestingBP'] <= (df['RestingBP'].mean() + 3*df['RestingBP'].std())]
df = df[ df['RestingBP'] >= (df['RestingBP'].mean() - 3*df['RestingBP'].std())]


df = df[ df['FastingBS'] <= (df['FastingBS'].mean() + 3*df['FastingBS'].std())]
df = df[ df['FastingBS'] >= (df['FastingBS'].mean() - 3*df['FastingBS'].std())]



df = df[ df['MaxHR'] <= (df['MaxHR'].mean() + 3*df['MaxHR'].std())]
df = df[ df['MaxHR'] >= (df['MaxHR'].mean() - 3*df['MaxHR'].std())]


df = df[ df['Oldpeak'] <= (df['Oldpeak'].mean() + 3*df['Oldpeak'].std())]
df = df[ df['Oldpeak'] >= (df['Oldpeak'].mean() - 3*df['Oldpeak'].std())]

```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

x[['ExerciseAngina','Sex','ChestPainType',"RestingECG","ST_Slope"]] = x[['ExerciseAngina','Sex','ChestPainType',"RestingECG","ST_Slope"]].apply(le.fit_transform)

x
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=30)
```

```python
from sklearn.decomposition import PCA

solo = PCA(0.95)

X_train = solo.fit_transform(X_train)
X_test = solo.fit_transform(X_test)
```

```python
from sklearn.svm import SVC

from sklearn.ensemble import RandomForestClassifier

from sklearn.linear_model import LogisticRegression

from sklearn.naive_bayes import GaussianNB,MultinomialNB

from sklearn.tree import DecisionTreeClassifier


sop = RandomForestClassifier()

sop.fit(X_train,y_train)


sop.score(X_train,y_train)
```

```python
model_parms = {
    'svm': {
        'model':SVC(gamma='auto'),
        'params':{
            'C':[1,10,20],
            'kernel':['rbf','linear']
        }
    },
    'random_forest': {
        'model':RandomForestClassifier(),
        'params':{
            
            'n_estimators':[1,5,10]
        }
    },
    'logistic_regression': {
        'model':LogisticRegression(solver = 'liblinear',multi_class = 'auto'),
        'params':{
            'C':[1,5,10]
        }
    },
    
    'decision_tree': {
        'model':DecisionTreeClassifier(),
        'params':{
            'criterion': ['gini','entropy'],
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV

scores = []

for m,k in model_parms.items():
   

    clf = GridSearchCV(k['model'],k['params'],cv=5,return_train_score = False)
    clf.fit(x,y)
    scores.append({'model':m,'best_scores':clf.best_score_,'best_params':clf.best_params_})
    
    

    


    
```
