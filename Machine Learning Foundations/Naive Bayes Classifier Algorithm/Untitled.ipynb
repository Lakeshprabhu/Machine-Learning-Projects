```python
import pandas as pd

df = pd.read_csv("spam.csv")
df
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x == "spam" else 0)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.Message,df.Category, test_size=0.25)


```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

xo = v.fit_transform(X_train.values)

```

```python
xo.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB

reg = MultinomialNB()

reg.fit(xo,y_train)
```

```python
reg.score(xo,y_train)
```

```python
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())

     ])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_train,y_train)
```

```python

from sklearn.datasets import load_wine

wine = load_wine()

dir(wine)
```

```python
wine.feature_names
```

```python
winedf = pd.DataFrame(wine.data, columns = wine.feature_names)
```

```python
winedf['type'] = wine.target
```

```python
winedf

X = winedf.drop('type',axis = 'columns')
```

```python
y = winedf['type']


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33)

```

```python
from sklearn.naive_bayes import GaussianNB

mod = GaussianNB()

mod.fit(X_train,y_train)
```

```python
mod.score(X_train,y_train)
```

```python
from sklearn.naive_bayes import MultinomialNB

mod1 = MultinomialNB()


mod1.fit(X_train,y_train)


```

```python
mod1.score(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(mod,X,y)
```

```python
cross_val_score(mod1,X,y)
```
