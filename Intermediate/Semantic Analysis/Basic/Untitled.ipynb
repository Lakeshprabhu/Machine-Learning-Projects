```python
import pandas as pd

df = pd.read_csv("redmi6.csv")
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.Rating = df.Rating.str.split(" ")
```

```python
df.Rating = df.Rating.apply(lambda x : x[0])
```

```python
df.Rating =(df.Rating).astype(float)
```

```python
df.Rating.dtype
```

```python
use_df = df[['Rating','Comments']]
```

```python
use_df
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC

clf = Pipeline([('vectorizer',CountVectorizer()),('sv',SVC())])
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(use_df.Comments,use_df.Rating,train_size = 0.8)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_train,y_train)
```

```python
po = clf.predict(X_test)
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_test,po))
```

```python

taco = pd.DataFrame({'predict':po,'actual':y_test})
check_list = []

for i in range(len(taco)):
    val1 = taco["predict"].iloc[i]
    val2 = taco["actual"].iloc[i]
    if val1==val2:
        check_list.append(1)
    else:
        check_list.append(0)

taco['check'] = check_list
    

```

```python


print(f"Percentage of correct {taco.check.value_counts().iloc[0] / len(taco)}")
```
