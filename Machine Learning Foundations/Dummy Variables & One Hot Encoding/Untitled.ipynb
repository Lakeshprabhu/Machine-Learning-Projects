```python
import pandas as pd

df = pd.read_csv("homeprices (2).csv")

df
```

```python
dummies = pd.get_dummies(df.town,dtype=int)
```

```python
df2 = pd.concat([df,dummies],axis='columns')
```

```python
df2
```

```python
df2 = df2.drop(['town','west windsor'],axis='columns')
```

```python
df2
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()

```

```python
X = df2.drop('price',axis='columns')
```

```python
reg.fit(X,df2.price)
```

```python
reg.predict([[2800,0,1]])
```

```python
reg.score(X,df2.price)
```
