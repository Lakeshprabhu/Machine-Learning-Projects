```python
sn = pd.read_csv("AB_NYC_2019.csv")
sn
```

```python
min1,max1 = sn.price.quantile([0.01,0.95])
min2,max2 = sn.minimum_nights.quantile([0.05,0.95])
min3,max3 = sn.availability_365.quantile([0.05,0.95])
min4,max4 = sn.number_of_reviews.quantile([0.05,0.95])
max5 = sn.reviews_per_month.quantile(0.95)
```

```python
sn = sn[(sn.price > min1) & (sn.price < max1)]
sn = sn[(sn.minimum_nights > min2) & (sn.minimum_nights < max2)]
sn = sn[(sn.number_of_reviews > min4) & (sn.number_of_reviews < max4)]
sn = sn[(sn.availability_365 > min3) & (sn.availability_365 < max3)]

sn = sn[(sn.availability_365 < max5)]
```
