```python
import pandas as pd

df = pd.read_csv("diabetes.csv")

df.describe()
```

```python
df.duplicated().any()
```

```python
df['Outcome'].value_counts()
```

```python
X = df.drop('Outcome',axis='columns')
```

```python
X
```

```python
y = df.Outcome
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

logo = StandardScaler()
x = logo.fit_transform(X)
x
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score


scores = cross_val_score(DecisionTreeClassifier(),X_train,y_train,cv=5)
```

```python
import numpy as np

np.mean(scores)
```

```python
from sklearn.ensemble import BaggingClassifier

mod = BaggingClassifier( DecisionTreeClassifier(),n_estimators =100,max_samples = 0.8, oob_score=True)

mod.fit(X_train,y_train)
```

```python
mod.oob_score_
```

```python
np.mean(cross_val_score(mod,X_train,y_train,cv=5))
```

```python
hdf = pd.read_csv("heart.csv")
hdf
```

```python
hdf.describe()
```

```python
hdf = hdf[hdf['Cholesterol'] < (hdf['Cholesterol'].mean() + 3*hdf['Cholesterol'].std())]
hdf = hdf[hdf['Cholesterol'] > (hdf['Cholesterol'].mean() - 3*hdf['Cholesterol'].std())]


hdf = hdf[hdf['RestingBP'] < (hdf['RestingBP'].mean() + 3*hdf['RestingBP'].std())]
hdf = hdf[hdf['RestingBP'] > (hdf['RestingBP'].mean() - 3*hdf['RestingBP'].std())]


hdf = hdf[hdf['MaxHR'] < (hdf['MaxHR'].mean() + 3*hdf['MaxHR'].std())]
hdf = hdf[hdf['MaxHR'] > (hdf['MaxHR'].mean() - 3*hdf['MaxHR'].std())]


hdf = hdf[hdf['Oldpeak'] < (hdf['Oldpeak'].mean() + 3*hdf['Oldpeak'].std())]
hdf = hdf[hdf['Oldpeak'] > (hdf['Oldpeak'].mean() - 3*hdf['Oldpeak'].std())]
```

```python
hdf
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

hdf[['ChestPainType','RestingECG','ExerciseAngina','ST_Slope','Sex']] = hdf[['ChestPainType','RestingECG','ExerciseAngina','ST_Slope','Sex']].apply(le.fit_transform)
```

```python
x = hdf.drop('HeartDisease',axis='columns')

y = hdf.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler

se = StandardScaler()

x = se.fit_transform(x)

```

```python
x
```

```python
from sklearn.svm import SVC

mod = SVC()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25)


mod.fit(X_train,y_train)

mod.score(X_train,y_train)
```

```python
sc = cross_val_score(mod,X_train,y_train,cv=5)

np.mean(sc)
```

```python
from sklearn.ensemble import BaggingClassifier

modo = BaggingClassifier( DecisionTreeClassifier(),n_estimators =100,max_samples = 0.8, oob_score=True)


modo.fit(X_train,y_train)
```

```python
modo.oob_score_
```

```python
np.mean(cross_val_score(modo,X_train,y_train,cv=5))
```
