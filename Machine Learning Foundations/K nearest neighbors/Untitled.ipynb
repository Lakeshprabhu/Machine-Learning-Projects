```python
import pandas as pd

from sklearn.datasets import load_iris 

iris = load_iris()

dir(iris)
```

```python
df = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
df['target'] = iris.target
```

```python
df['t2'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
df1 = df[df['target'] == 0]
df2 = df[df['target'] == 1]
df3 = df[df['target'] == 2]
```

```python
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='red')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='blue')
plt.scatter(df3['sepal length (cm)'],df3['sepal width (cm)'],color='green')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['t2'],axis='columns')
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier

rog = KNeighborsClassifier(n_neighbors=3)

rog.fit(X_train,y_train)
```

```python
rog.score(X_train,y_train)
```

```python
pred = rog.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,pred))
```

```python
from sklearn.datasets import load_digits

data = load_digits()


xo_train, xo_test, yo_train, yo_test = train_test_split(
    data.data, data.target, test_size=0.2)
```

```python
rogo = KNeighborsClassifier(n_neighbors=3)

rogo.fit(xo_train,yo_train)
```

```python
pop = rogo.predict(xo_test)
```

```python

print(classification_report(yo_test,pop))
```
