```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df['area'],df['price'],color ='red',marker='+')
```

```python
reg = linear_model.LinearRegression()

```

```python
x=df[['area']]
y = df.price

```

```python
reg.fit(x,y)
```

```python
reg.predict([[3000]])
```

```python
areas_to_predict = pd.read_csv("areas.csv")

areas_to_predict
```

```python
p = reg.predict(areas_to_predict)
```

```python
areas_to_predict['price'] = p
```

```python
areas_to_predict.to_csv("prediction.csv",index=False)
areas_to_predict
```

```python
%matplotlib inline

plt.xlabel("area")
plt.ylabel('price')
plt.scatter(areas_to_predict['area'],reg.predict(areas_to_predict[['area']]),color='green',marker='*')
```
