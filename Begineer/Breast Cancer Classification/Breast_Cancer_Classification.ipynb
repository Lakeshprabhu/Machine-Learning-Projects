```python
import pandas as pd
import numpy as np

import seaborn as sns

from sklearn.datasets import load_breast_cancer
```

```python
c = load_breast_cancer()

c.keys()
```

```python
df = pd.DataFrame(c.data,columns = c.feature_names)
df
```

```python
df['target']=c.target
```

```python
df
```

```python
sns.pairplot(df,vars=['mean radius','mean texture','mean area','mean perimeter','mean smoothness'])
```

```python
df.columns
```

```python
sns.scatterplot(x='mean area',y='mean smoothness',hue='target',data = df)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('target',axis='columns')
y = df.target

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8)
```

```python
from sklearn.svm import SVC

model = SVC()

model.fit(X_train,y_train)
```

```python
po = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report

co = confusion_matrix(y_test,po)

print(classification_report(y_test,po))
```

```python
sns.heatmap(co,annot=True)
```

```python
from sklearn.preprocessing import StandardScaler

i = StandardScaler()

x = i.fit_transform(X)


X_train,X_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

```python
modo2 = SVC()

modo2.fit(X_train,y_train)

po  = modo2.predict(X_test)
```

```python
print(classification_report(y_test,po))
```

```python
param_grid={'C':[0.1,1,10,100],"gamma":[1,0.1,0.01,0.001]}
```

```python
 from sklearn.model_selection import GridSearchCV

 cv = GridSearchCV(SVC(),param_grid,refit = True,verbose=4, return_train_score=False)
```

```python
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
pop = cv.predict(X_test)
```

```python
coz = confusion_matrix(y_test,pop)
```

```python
sns.heatmap(coz,annot=True)
```

```python
print(classification_report(y_test,pop))
```
